```python
import tkinter as tk

class Calculator:
    def __init__(self, master):
        self.master = master
        master.title("Calculadora")

        # Set window size for vertical mobile resolution
        # You can adjust these values for specific phone resolutions
        master.geometry("360x640")  # e.g., common mobile portrait size
        master.resizable(False, False) # Disable resizing

        self.equation = tk.StringVar()
        self.equation.set("0")

        # Entry widget to display the current equation
        self.display = tk.Entry(master, textvariable=self.equation, font=('Arial', 24), bd=10, insertwidth=2, width=14, borderwidth=4, justify='right')
        self.display.grid(row=0, column=0, columnspan=4, pady=10, padx=10, sticky="nsew")

        # Define button layout
        buttons = [
            ('C', 1, 0), ('/', 1, 1), ('*', 1, 2), ('DEL', 1, 3),
            ('7', 2, 0), ('8', 2, 1), ('9', 2, 2), ('-', 2, 3),
            ('4', 3, 0), ('5', 3, 1), ('6', 3, 2), ('+', 3, 3),
            ('1', 4, 0), ('2', 4, 1), ('3', 4, 2), ('=', 4, 3),
            ('0', 5, 0), ('.', 5, 1)
        ]

        # Configure rows and columns to expand proportionally
        for i in range(6):
            master.grid_rowconfigure(i, weight=1)
        for i in range(4):
            master.grid_columnconfigure(i, weight=1)

        # Create buttons
        for (text, row, col) in buttons:
            if text == '=':
                button = tk.Button(master, text=text, font=('Arial', 18), command=self.evaluate_expression)
                button.grid(row=row, column=col, columnspan=2, sticky="nsew", padx=5, pady=5)
            elif text == '0':
                button = tk.Button(master, text=text, font=('Arial', 18), command=lambda t=text: self.button_press(t))
                button.grid(row=row, column=col, columnspan=2, sticky="nsew", padx=5, pady=5)
            else:
                button = tk.Button(master, text=text, font=('Arial', 18), command=lambda t=text: self.button_press(t))
                button.grid(row=row, column=col, sticky="nsew", padx=5, pady=5)
        
        # Adjusting the '=' button to span two columns if '0' also spans two
        # In this layout, '=' spans 2 columns, and '0' spans 2 columns,
        # so '.' must be at 5,1
        zero_button = tk.Button(master, text='0', font=('Arial', 18), command=lambda: self.button_press('0'))
        zero_button.grid(row=5, column=0, columnspan=2, sticky="nsew", padx=5, pady=5)

        dot_button = tk.Button(master, text='.', font=('Arial', 18), command=lambda: self.button_press('.'))
        dot_button.grid(row=5, column=2, sticky="nsew", padx=5, pady=5)

        equal_button = tk.Button(master, text='=', font=('Arial', 18), command=self.evaluate_expression)
        equal_button.grid(row=4, column=3, rowspan=2, sticky="nsew", padx=5, pady=5) # spans 2 rows for better mobile layout

    def button_press(self, char):
        current_equation = self.equation.get()
        if current_equation == "0" and char not in ['C', 'DEL', '/', '*', '-', '+', '.']:
            self.equation.set(char)
        elif char == 'C':
            self.equation.set("0")
        elif char == 'DEL':
            if len(current_equation) > 1:
                self.equation.set(current_equation[:-1])
            else:
                self.equation.set("0")
        else:
            self.equation.set(current_equation + char)

    def evaluate_expression(self):
        try:
            result = str(eval(self.equation.get()))
            self.equation.set(result)
        except Exception as e:
            self.equation.set("Erro")

if __name__ == "__main__":
    root = tk.Tk()
    calculator = Calculator(root)
    root.mainloop()
```
